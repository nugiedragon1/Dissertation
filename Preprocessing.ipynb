```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
df1 = pd.read_csv("bmw.csv")
df2 = pd.read_csv("audi.csv")
df3 = pd.read_csv("toyota.csv")
df4 = pd.read_csv("merc.csv")
df5 = pd.read_csv("vw.csv")

frames = [df1,df2,df3,df4,df5]
df=pd.concat(frames)
# df.isna().sum()
df.reset_index(inplace=True,drop=True)
df
```

```python
df.isna().sum()
```

```python
df_predictors = df1.drop(['price'], axis = 1)
x = df_predictors.select_dtypes(exclude=['object'])
y=df1.price
```

```python
rfe = RFE(estimator=DecisionTreeClassifier(),n_features_to_select=2)

rfe.fit(x,y)

for i in range(x.shape[1]):
    print('Column: %s, Selected=%s, Rank: %d' % (x.columns[i],rfe.support_[i],rfe.ranking_[i]))
```

```python
scaler = MinMaxScaler()

x_scaled = pd.DataFrame(scaler.fit_transform(x))

x_scaled
```

```python
trans = PCA(n_components = 2)

df_pca=pd.DataFrame(trans.fit_transform(x_scaled))

df_pca
```
